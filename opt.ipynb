```python
import numpy as np


def consistent(A, b):
    """Check Ax=b feasibility."""
    if len(A) == 0:
        return True

    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float).reshape(-1, 1)

    return np.linalg.matrix_rank(A) == np.linalg.matrix_rank(
        np.hstack([A, b])
    )


def feasible(constraints, A=[], b=[], i=0):
    """
    constraints:
        [(a,b,c,d), ...]
        representing:
        (a^T x + b)(c^T x + d)=0
    """

    if i == len(constraints):
        return True

    a, b0, c, d = constraints[i]

    # choose a^T x + b0 = 0
    if consistent(A + [a], b + [-b0]):
        if feasible(constraints, A + [a], b + [-b0], i+1):
            return True

    # choose c^T x + d = 0
    if consistent(A + [c], b + [-d]):
        if feasible(constraints, A + [c], b + [-d], i+1):
            return True

    return False


# (x1)(x2)=0
# (x1+x2-1)(x1-x2)=0
constraints = [
    (np.array([1,0]), 0, np.array([0,1]), 0),
    (np.array([1,1]), -1, np.array([1,-1]), 0),
    (np.array([1,2]), 1, np.array([1,1]), 1)
]

print(feasible(constraints))
```
